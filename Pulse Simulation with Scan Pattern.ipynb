# Simulation of received pulses from uniformly illuminated rectangular aperture radar

### Based on "MODELING OF A RADAR SIGNAL FOR SCAN PATTERN" (IEEE)
https://ieeexplore.ieee.org/document/5379909

```python
import numpy as np
import matplotlib.pyplot as plt

epsilon = 1e-9  # Small value to prevent division by zero

# Parameters
A = 1         # Amplitude scaling factor
b = 1.0         # Control parameter for the main lobe width
c = 10           # Frequency scaling factor
theta = np.linspace(-np.pi, np.pi, 1000)  # Angle range
k1 = 1e-1            # For main lobe shape
k2 = 1e-3            # Simulation of received pulses from uniformly illuminated rectangular aperture radar
k3 = 2e-1            # For no. of side lobes & directivity of antenna beam
el = epsilon        # Elevation

# Calculate the radiation pattern
x = np.sin(theta)
scan_pattern = np.sin(k1 * np.sin(theta/k3) * np.cos(el/k3)) * np.sin(k2 * np.sin(el/k3) * np.cos(theta/k3))
scan_pattern = scan_pattern / (k1 * np.sin(theta/k3) * np.cos(theta/k3)) / (k2 * np.sin(el/k3) * np.cos(theta/k3))
scan_pattern = scan_pattern**2

# Convert to dB scale
scan_pattern_db = 20 * np.log10(scan_pattern)

# Plot the results
plt.figure(figsize=(8, 4))
plt.plot(np.degrees(theta), scan_pattern)
plt.title(r'Power')
plt.xlabel('Angle (degrees)')
plt.ylabel('Power Received')
plt.ylim(0, 100000)
plt.xlim(-180, 180)
plt.grid(True)
plt.show()
```
